## Caractères de commande

```python
titre2 = 'résultat d\'expérience'
print (titre2)
```

```python
titre3 = "résultat d'expérience\n"
print (titre3)
```

Attention aux chemins d'accès !

```python
print('C:\some\path\name\PH461')
print(r'C:\some\path\name\PH461')
```

```python
titre4 = r"résultat d'expérience\n"
print (titre4)
```

## Majuscules

```python
titre3_uppercase = titre3.upper ()
print (titre3_uppercase)
```

```python
titre3_lowercase = titre3_uppercase.lower ()
print (titre3_lowercase)
```

```python
titre3bis = titre3_lowercase [0].upper () + titre3_lowercase [1:]
print (titre3bis)
```

```python
titre4 = titre3_lowercase.capitalize ()
print (titre4)
```

## Concaténation de chaînes de caractères

```python
titre5 = titre4 + '=' * len (titre3) + '\n'
print (titre5)
```

```python
titre6 = titre5 [:8] + 's' + titre5 [8:]
print (titre6)
```

## Replace()
L'instruction _replace_ permet de remplacer un caractère ou une chaîne de caractère par un autre caractère ou chaîne de caractère

```python
titre6bis = titre6.replace (' ', '_')
print (titre6bis)
print (titre6)
```

## Split()

La méthode _split()_ découpe une chaîne de caractères en plusieurs éléments appelés champs, en utilisant n'importe quel séparateur.

```python
titre6.split (' ')
```

```python
titre6.split ('é')
```

```python
titre6.split ('=')
```

La méthode `.find()`, quant à elle, recherche une chaîne de caractères passée en argument. 
Si l'élément recherché est trouvé, alors l'indice du début de l'élément dans la chaîne de caractères est renvoyé. 

```python
titre6.find ('s')
```

Si l'élément n'est pas trouvé, alors la valeur -1 est renvoyée.

```python
titre6.find ('X')
```

Si l'élément recherché est trouvé plusieurs fois, seul l'indice de la première occurrence est renvoyé

```python
titre6.find ('=')
```

## Count()

La méthode _count()_ compte le nombre d’occurrences d'une chaîne de caractères passée en argument

```python
titre6.count ('s')
```

```python
titre6.count ('X')
```

```python
titre6.count ('=')
```
